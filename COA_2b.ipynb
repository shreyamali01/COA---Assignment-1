```python
import numpy as np
#all the elements in the matrix are double
nums = [64,128,256,512,1024]
for i in range(len(nums)):
  p = nums[i]

  #initializing the matrices
  M1 = np.random.uniform(0,10,size=(p,p))
  M2 = np.random.uniform(0,9,size=(p,p))

  #initializing the resultant matrix with zeros
  ans = np.zeros((p,p),dtype=float)

  #matrix multiplication
  for i in range(p):
    for j in range(p):
      for k in range(p):
        ans[i,j] += M1[i,k] * M2[k,j]


  print(M1,'\n')
  print(M2, '\n')
  print(ans, '\n')
```
